```python
from corus import load_wiki
import re
from tqdm import tqdm
from nltk import tokenize
```

```python
path = 'ruwiki-latest-pages-articles.xml.bz2'
```

```python
cur_bad_word = "типа"
total_good_sentences = []
records = load_wiki(path)
```

```python
def filter_sentence(text):
    text = text.replace('\xa0', ' ')
    text = text.replace('=', '')
    text = text.replace('\xad', '')
    return text
```

```python
def get_sent_from_text(text, bad_word):
    good_sentences = []
    sents = tokenize.sent_tokenize(text)
    for sent in sents:
        sent = filter_sentence(sent)
        words_in_sent = re.split('\W+', sent)
        if bad_word in words_in_sent:
            good_sentences.append(sent)
    return good_sentences
```

```python
# len(records)
```

```python
cnt = 0
for rec in tqdm(records):
    cnt += 1
#     if cnt > 10000:
#         break
    words_in_sent = re.split('\W+', rec.text)
    if cur_bad_word in words_in_sent:
        total_good_sentences += get_sent_from_text(rec.text, cur_bad_word)
```
